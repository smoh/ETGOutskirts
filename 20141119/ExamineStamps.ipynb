We examine whether our image quality cuts introduces any bias in the sample. The following three plots who the distribution of the parent sample (blue) and the final sample (red) of $z$, $M_r$, and $\sigma$ on the upper panel, and the fractional difference in each bin on the lower panel. Nearby bright galaxies are preferentially exluded from the parent sample.

```python
def plot_hist_diff(d, true=None, **hist_kwargs):
    fig, (ax, ax_d) = subplots(2, 1, sharex=True)
    c, bins, patches = ax.hist(d, **hist_kwargs)
    c_sub, bins, patches = ax.hist(d[true], bins=bins)
    ax_d.bar(bins[:-1], (c_sub-c)/c, bins[1]-bins[0])
    return fig
```

We fit our galaxies with a deV+Exp two component model only holding the indices fixed.

We flag galaxies with reasonable fit parameters ($R_e < 80$ arcsec) and a dominant exponential outer component.
About half of the sample is flagged as having double component.

```python
def Rcross(p1, p2):
    x1, y1 = p1
    x2, y2 = p2
    y1[y1<0.001] = 0
    y2[y2<0.001] = 0
    d = y1-y2
    if any(d<0):
        return x1[d<0][0]
    else:
        return -1
```

```python
Rc = [Rcross(*(s.get_profile_comps(i, 'deVExp'))) for i in range(len(s.table))]
Rc = atleast_1d(Rc)
```

```python
m = s.models['deVExp']
cutValidFit = (m.p[1].Re>0) & (m.p[1].Re<2e2) & (m.p[0].Ie>0)
cutIntersect = Rc>0
cutDoubleComp = cutValidFit & cutIntersect & cutSample
ind1 = ind[cutDoubleComp]
print 'N_cutDoubleComp = %i (%.2f)' % (ind1.size, ind1.size/float(len(s.table)))
```

```python
da = bd.cosmo.angular_diameter_distance(s.table['Z_1'].data).value*1000/206265.
```

Figure below shows the correlation matrix between key parameters. Subscript 4 and 1 corresponds to deV and Exp component.
Double component galaxies are marked red.

```python
rc('mathtext', default='regular')
def d(model):
    return ma.masked_invalid(vstack([
        model.p[0].mu_e_avg,
        log10(model.p[0].Re*.396 * da),
        model.p[0].q,
        model.p[1].mu_e_avg,
        log10(model.p[1].Re*.396 * da),
        model.p[1].q,
        s.table['VDISP'].data,
        model.p.flux_ratios[0] ]).T)

fig = corner(d(s.models['deVExp']),
       labels=['$<\mu>_{e,4}$', 'log $R_{e,4}$/kpc', '$q_4$',
               '$<\mu>_{e,1}$', 'log $R_{e,1}$/kpc', '$q_1$', '$\sigma$ / km/s', 'B/T'],
       bins=20,
       plot_datapoints=True, plot_contours=False,
       mark_bool=cutDoubleComp,
       plot_kwargs=dict(ms=3))
```

- $B/T$ ranges from 0.2 to 0.9 with a median of 0.6.
- Median ratio of effective radius for double component galaxies is 6.

```python
da = bd.cosmo.angular_diameter_distance(s.table['Z_1'].data).value*1000/206265.
devMag, expMag = m.p[0].flux_mag, m.p[1].flux_mag
devMag_corr = devMag - s.table['EXTINCTION'][:,4]-s.table['KCORRECT'][:,4]
expMag_corr = devMag - s.table['EXTINCTION'][:,4]-s.table['KCORRECT'][:,4]
devRo_kpc, expRo_kpc = m.p[0].Ro*.396*da, m.p[1].Ro*.396*da

devRFP_kpc = bd.Bernardi03_r.R0(s.table['VDISP'], devMag, s.table['Z_1'])
devRFP_kpc_corr = bd.Bernardi03_r.R0(s.table['VDISP'], devMag_corr, s.table['Z_1'])
expRFP_kpc = bd.Bernardi03_r.R0(s.table['VDISP'], expMag, s.table['Z_1'])
expRFP_kpc_corr = bd.Bernardi03_r.R0(s.table['VDISP'], expMag_corr, s.table['Z_1'])
```

We check if there is any FP-like relation for each component of double component galaxies (larger symbol)
using Bernerdi et al. 2003 FP coefficients.
The two plots are identical with deV component highlighted on the left and Exp eomponent highlighted on the right.
Smaller symbols correspond to Single component galaxies.

Horizontal and vertical dashed lines mark 1.5 arcsec (~size of the PSFs) at a median redshift of .025.

```python
fig, ax = subplots(1, 2, figsize=(12, 4), sharex=True, sharey=True)

def pp(alpha1, alpha2):
    scatter(devRo_kpc, devRFP_kpc_corr, c='r', alpha=alpha1)
    if alpha1 is 1:
        scatter(devRo_kpc[cutDoubleComp], devRFP_kpc_corr[cutDoubleComp], c='r',
                s=80)#, facecolor='None', edgecolor='k')
    scatter(expRo_kpc, expRFP_kpc_corr, c='b', alpha=alpha2)
    if alpha2 is 1:
        scatter(expRo_kpc[cutDoubleComp], expRFP_kpc_corr[cutDoubleComp],
                s=80)#, facecolor='None', edgecolor='r')
    xscale('log')
    yscale('log')
    x = 10**arange(-1., 2., .1)
    plot(x,x, 'k-', lw=1)
    ylim(3e-2, 20)
    xlabel('Component R0 kpc')
    ylabel('Component R_FP kpc')
    
    axhline(1.5 * bd.cosmo.angular_diameter_distance(.025).value /206.265, ls='--', c='k', lw=1)
    axvline(1.5 * bd.cosmo.angular_diameter_distance(.025).value /206.265, ls='--', c='k', lw=1)
sca(ax[0])
pp(1, .3)
sca(ax[1])
pp(.3, 1)

```
