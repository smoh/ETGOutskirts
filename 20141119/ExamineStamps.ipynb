```python
def plot_hist_diff(d, true=None, **hist_kwargs):
    fig, (ax, ax_d) = subplots(2, 1, sharex=True)
    c, bins, patches = ax.hist(d, **hist_kwargs)
    c_sub, bins, patches = ax.hist(d[true], bins=bins)
    ax_d.bar(bins[:-1], (c_sub-c)/c, bins[1]-bins[0])
    return fig
```

We fit our galaxies with a deV+Exp two component model only holding the indices fixed.

We flag galaxies with reasonable fit parameters ($R_e < 80$ arcsec) and a dominant exponential outer component.
About half of the sample is flagged as having double component.

```python
def Rcross(p1, p2):
    x1, y1 = p1
    x2, y2 = p2
    y1[y1<0.001] = 0
    y2[y2<0.001] = 0
    d = y1-y2
    if any(d<0):
        return x1[d<0][0]
    else:
        return -1
```

```python
Nbins = r_[0, 3, 10, 1e3]
cNnormed, cN = [], []
be = r_[9., 10., 11., 12.]
bc = (be[1:] + be[:-1])*0.5
for i, cc in enumerate(bc):
    inbin = imassbin == i
    c, _ = histogram(tg['groupN'][inbin & cutDoubleComp], bins=Nbins)
    cc, _ = histogram(tg['groupN'][inbin], bins=Nbins)
    cN.append(c.astype(float))
    cNnormed.append(c.astype(float) / cc.astype(float))

ax = subplot(111)
ax_twin = ax.twinx()
for val, valn, mass in zip(cN, cNnormed, bc):
    ax_twin.plot([1,2,3], val, 'x--', label='%.1f'%(mass))
    ax.plot([1,2,3], valn, 'o-', label='%.1f'%(mass))
xlim(0.5,3.5)
xticks([1,2,3], ['<3','3 < N < 10', '>10'])
ax.set_ylabel('frac double')
ax.set_xlabel('group N bins')
ax_twin.set_ylabel('count')
legend()
```
