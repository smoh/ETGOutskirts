```python
from astropy.table import Table
%pylab inline
```

```python
vis = Table.read('visualflag.dat', format='ascii', names=['vflag'])
master = Table.read('sample.fits')
```

```python
figure()
n, bins, patches = hist(
    master['Z_1'].data, histtype='step', label='entire')
n1, bins, patches1 = hist(
    master['Z_1'].data[vis['vflag']=='0'], bins=bins, histtype='step', label='good')
legend(loc='upper left')
```

```python
figure()
xlabel('redshift')
ylabel('sigma (km/s)')
scatter(master['Z_1'].data, master['VDISP'].data, c='0.5', edgecolors='None')
scatter(master['Z_1'].data[vis['vflag']=='0'], master['VDISP'][vis['vflag']=='0'].data)
```

```python
mag_r = 22.5-2.5*log10(master['SERSICFLUX'][:,4].data)
sigma = master['VDISP'].data
Re = master['SERSIC_TH50'].data

good = vis['vflag']=='0'

figure()
xlabel('sigma (km/s)')
ylabel('mag')
scatter(sigma, mag_r, c='0.5')
scatter(sigma[good], mag_r[good], c='r')

figure()
xlabel('sigma (km/s)')
ylabel('Re')
yscale('log')
scatter(sigma, Re, c='0.5')
scatter(sigma[good], Re[good], c='r')

figure()
xlabel('mag')
ylabel('Re')
yscale('log')
scatter(mag_r, Re, c='0.5')
scatter(mag_r[good], Re[good], c='r')
```
