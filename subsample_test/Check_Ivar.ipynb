## My sdss images reprojected to NSA to use deblending info

```python
iauname = 'J080720.25+510753.9'
data = fits.getdata('data/deblended/'+iauname+'.fits')
ivar = fits.getdata('data/ivar/'+iauname+'.fits')
model = fits.getdata('sdss_psf/dvc/deblended/models/M'+iauname+'.fits')
```

```python
figure()
imshow(log10(data))
figure()
imshow(ivar)
```

```python
# background region
dsub = data[600:700,200:400].copy().ravel()
ivsub = ivar[600:700,200:400].copy().ravel()
good = dsub != 0

print std(dsub[good])
print median(1./sqrt(ivsub[good]))
```

```python
sig = dsub[good] * sqrt(ivsub[good])
hist(sig, bins=50, normed=True) #, weights=ones_like(sig)/len(sig) )
x = arange(-4.,4.,0.01)
plot(x, exp(-x**2/2)/sqrt(2*pi),'r')
print mean(sig), std(sig)
print sum(sig**2)/(len(sig)-7)
figure()
hist(sig**2, bins=30, cumulative=True, normed=True)
```

```python
ivar[data==0] = 0.
print sum(ivar*(data-model)**2)/count_nonzero(data!=0)
```

## SDSS frame images

```python
d = fits.getdata('data/sdss_field/frame-r-000756-1-0546.fits')
iv = fits.getdata('data/sdss_ivar/frame-r-000756-1-0546.fits')
figure()
imshow(log10(d[1000:1300,1000:1400]))
figure()
imshow(iv[1000:1300,1000:1400])
```

```python
ds = d[1000:1300,1000:1200].copy()
ivs = iv[1000:1300,1000:1200].copy()
print std(ds)
print median(1./sqrt(ivs))
```

```python
s = ds.ravel() * sqrt(ivs.ravel())
print mean(s)
hist(s, bins=50, normed=True)
x = arange(-4.,4.,0.01)
plot(x, exp(-x**2/2)/sqrt(2*pi),'r')
```
