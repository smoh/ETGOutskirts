```python
%pylab inline
from pybdfitter.fp import Bernardi03_r
from pybdfitter.sersic import NSersic
from pybdfitter.plot import Plotter
from astropy.table import Table
from astropy import cosmology
from triangle import corner
import mpltools
mpltools.set_ipynb()
cosmo = cosmology.FlatLambdaCDM(70., 0.3)
```

```python
dvc = Table.read('fit/dvc/RAWFIT00000.00268.fits')
dvcexp = Table.read('fit/dvcexp/RAWFIT00000.00268.fits')
m = Table.read('sample.fits')
v = Table.read('visualflag.dat', format='ascii', names=['vflag'])
good = v['vflag']=='0'

sigma = m['VDISP'].data
z = m['Z_1'].data
kcorr = m['KCORRECT'][:,4].data
ext = m['EXTINCTION'][:,4].data
DA = cosmo.angular_diameter_distance(z).value*1e3/206265.
```

```python
pDVC1 = NSersic(dvc['FIT_DVC'].data.T)
I0 = (pDVC1[0].mu_e_avg - 10*log10(1+z) - kcorr - ext)/-2.5
R0_fp = 10**(Bernardi03_r.a*log10(sigma) + Bernardi03_r.b*I0 + Bernardi03_r.c)
R0_fit = pDVC1[0].Ro*0.396*DA

pDVCEXP = NSersic(dvcexp['FIT_DVCEXP'].data.T)
pDVC = pDVCEXP[0]
pEXP = pDVCEXP[1]

DVC_I0 = (pDVC.mu_e_avg - 10*log10(1+z) - kcorr - ext)/-2.5
DVC_R0_fp = 10**(Bernardi03_r.a*log10(sigma)+Bernardi03_r.b*DVC_I0+Bernardi03_r.c)
DVC_R0_fit = pDVC.Ro*0.396*DA

EXP_I0 = (pEXP.mu_e_avg - 10*log10(1+z) - kcorr - ext)/-2.5
EXP_R0_fp = 10**(Bernardi03_r.a*log10(sigma)+Bernardi03_r.b*DVC_I0+Bernardi03_r.c)
EXP_R0_fit = pEXP.Ro*0.396*DA
```

```python
figure(figsize=(15,10))
subplot(221)
xlabel('R0 DVC')
ylabel('R0 DVCEXP')
scatter(log10(R0_fit), log10(DVC_R0_fit),
        c='r', alpha=0.8, label='DVCEXP DVC Comp')
scatter(log10(R0_fit), log10(EXP_R0_fit),
        marker='+', c='b', label='DVCEXP EXP Comp')
scatter([log10(R0_fit)[9]], [log10(DVC_R0_fit)[9]],
        s=150, marker='+', edgecolors='k', label='J123948.37')
legend(loc='upper left')
# identity line
a=arange(-1,2,0.1)
plot(a,a,'r')
xlim(-1, 2)
ylim(-1,3.5)

subplot(222)
xlabel('R0 FP')
ylabel('R0 FIT')
scatter(log10(EXP_R0_fp), log10(EXP_R0_fit),
        marker='+', edgecolors='b', s=50, label='DVCEXP DVC Comp')
scatter(log10(DVC_R0_fp), log10(DVC_R0_fit),
        c='r', s=50, alpha=0.6, label='DVCEXP EXP Comp')
scatter(log10(R0_fp), log10(R0_fit),
        c='0.5', label='DVC')
scatter([log10(DVC_R0_fp)[9]], [log10(DVC_R0_fit)[9]],
        s=150, c='m', label='J123948.37 DVCEXP DVC Comp')
scatter([log10(EXP_R0_fp)[9]], [log10(EXP_R0_fit)[9]],
        s=150, c='k', marker='^', label='J123948.37 DVCEXP EXP Comp')
legend(loc='upper left')
# identity line
a=arange(-1,2,0.1)
plot(a,a,'r')
xlim(-1,2)
ylim(-1,3.5)

subplot(223)
ylabel('delta $\chi^2$')
xlabel('ID')
plot(dvcexp['CHISQ_DVCEXP']*dvcexp['DOF_DVCEXP']-dvc['CHISQ_DVC']*dvc['DOF_DVC'], 'ko')
yscale('symlog')
```
