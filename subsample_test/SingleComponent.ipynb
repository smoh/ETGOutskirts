```python
%pylab inline
from astropy.table import Table
from pybdfitter.sersic import Sersic
from pybdfitter.fp import Bernardi03_r
from astropy import cosmology
import triangle
import mpltools
mpltools.set_ipynb()
cosmo = cosmology.FlatLambdaCDM(70., 0.3)
```

```python
master = Table.read('sample.fits')  # NSA+morph
dvc = Table.read('fit/dvc/RAWFIT00000.00268.fits')
ser = Table.read('fit/ser/RAWFIT00000.00268.fits')
```

```python
z = master['Z_1'].data
kcorr = master['KCORRECT'][:,4].data
ext = master['EXTINCTION'][:,4].data
DA = cosmo.angular_diameter_distance(z).value * 1000./206265.  # kpc/arcsec
sigma = dvc['VDISP'].data
chisq = dvc['CHISQ_DVC'].data
# DVC
pdvc = Sersic(dvc['FIT_DVC'].data.T)
I0 = (pdvc.mu_e_avg - kcorr - ext - 10*log10(1+z))/-2.5
R0_kpc = pdvc.Ro * 0.396 * DA
R0_fp = 10**(Bernardi03_r.a*log10(sigma) + Bernardi03_r.b*I0 + Bernardi03_r.c)
# SER
pser = Sersic(ser['FIT_SER'].data.T)
```

### Classical Fundamental Plane of single De Vaucoulerus profile

```python
figure()
diff_dex = log10(R0_kpc) - log10(R0_fp)
offset, scat = median(diff_dex), std(diff_dex)
print 'offset =', offset, 'scatter =', scat
scatter(log10(R0_fp), log10(R0_kpc), c=sigma, edgecolors='None')
a = linspace(-0.4, 1.2, 50)
plot(a,a,'r', a, a+offset, 'r--')
xlim(-0.4,1.2)
xlabel('log R0 Bernardi+03')
ylabel('log R0 fit')
cb = colorbar()
cb.ax.set_ylabel('sigma km/s', rotation=-90, labelpad=20)
```

```python
d = vstack((I0, log10(R0_kpc), sigma, chisq, pdvc.q)).T
a = triangle.corner(d, labels=["I0", "log R0", "sigma", "chisq", "q"], plot_contours=False)
```

### Comparison with Sersic profile:
in order to put more light at the center or at the outer part,

- Sersic index will be higher than de Vaucouleurs
- effective radius will increase

```python
figure()
hist(pser.n, histtype='step', bins=20)
axvline(4, c='r')
xlabel('Sersic n')
```

```python
figure(figsize=(15,6))
subplot(121)
xscale('log')
yscale('log')
xlabel('Re DVC pixel')
ylabel('Re SER pixel')
chi_delta = (ser['CHISQ_SER']*ser['DOF_SER'] - dvc['CHISQ_DVC']*dvc['DOF_DVC']).data
scatter(pdvc.Re, pser.Re, c=log10(abs(chi_delta)))
colorbar().ax.set_ylabel('log delta $\chi^2$', rotation=-90, labelpad=20)
a = arange(1,100)
plot(a,a,c='r')
xlim(1,100)
ylim(1,300)

subplot(122)
hist(log10(abs(chi_delta)))
xlabel('log delta $\chi^2$')
```

```python
d = vstack(( sigma, pser.mu_e_avg/-2.5, pser.n, log10(pser.Re), pser.q)).T
a = triangle.corner(d, labels=['sigma', 'I0', 'n', 'Re', 'q'], plot_contours=False)
```

```python
I0_ser = (pser.mu_e_avg - kcorr - ext - 10*log10(1+z))/-2.5
R0_kpc_ser = pser.Ro * 0.396 * DA
R0_fp_ser = 10**(Bernardi03_r.a*log10(sigma) + Bernardi03_r.b*I0_ser + Bernardi03_r.c)

figure()
diff_dex = log10(R0_kpc_ser) - log10(R0_fp_ser)
offset, scat = median(diff_dex), std(diff_dex)
print 'offset =', offset, 'scatter =', scat
scatter(log10(R0_fp_ser), log10(R0_kpc_ser), label='Sersic')
scatter(log10(R0_fp), log10(R0_kpc), c='0.5', label='De Vaucouleurs')
a = linspace(-0.4, 2, 50)
plot(a,a,'r', a, a+offset, 'r--')
xlim(-0.4,1.7)
ylim(-0.5, 1.7)
xlabel('log R0 Bernardi+03')
ylabel('log R0 fit')
legend(loc='upper left')


figure()
scatter(log10(R0_fp_ser), log10(R0_kpc_ser), c=log10(abs(chi_delta)))
a = linspace(-0.4, 1.7, 50)
plot(a,a,'r', a, a+offset, 'r--')
xlim(-0.4,1.7)
ylim(-0.5, 1.7)
xlabel('log R0 Bernardi+03')
ylabel('log R0 fit')
colorbar().ax.set_ylabel("log delta $\chi^2$", rotation=-90, labelpad=20)
```
