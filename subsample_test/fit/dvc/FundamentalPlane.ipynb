```python
%pylab inline
from astropy.table import Table
from pybdfitter.sersic import Sersic
from pybdfitter.fp import Bernardi03_r
from astropy import cosmology
cosmo = cosmology.FlatLambdaCDM(70., 0.3)
```

```python
import mpltools
mpltools.set_pub_single()
```

```python
master = Table.read('../../sample.fits')
z = master['Z_1'].data
result = Table.read('RAWFIT00000.00268.fits')
p = Sersic(result['FIT_DVC'].data.T)
sigma = result['VDISP'].data
I0 = p.mu_e_avg/-2.5
chisq = result['CHISQ_DVC'].data
```

```python
DA = cosmo.angular_diameter_distance(z).value * 1000./206265.
R0_kpc = p.Ro * 0.396 * DA
```

```python
R0_fp = 10**(Bernardi03_r.a*log10(sigma) + Bernardi03_r.b*I0 + Bernardi03_r.c)
diff_dex = log10(R0_kpc) - log10(R0_fp)
offset, scat = median(diff_dex), std(diff_dex)
print offset, scat
scatter(log10(R0_fp), log10(R0_kpc), c=log10(chisq), edgecolors='None')
a = linspace(-0.4, 1.2, 50)
plot(a,a,'r', a, a+offset, 'r--')
xlim(-0.4,1.2)
xlabel('log R0_Bernardi+03')
ylabel('log R0_fit')
colorbar()
```

```python
hist(R0_kpc/R0_fp, bins=20)
xlabel('R0_fit/R0_B03')
```
