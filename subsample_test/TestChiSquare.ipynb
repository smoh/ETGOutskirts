```python
dvc = Table.read('fit/dvc/RAWFIT00000.00268.fits')
ser = Table.read('fit/ser/RAWFIT00000.00268.fits')
dvcexp = Table.read('fit/dvcexp/RAWFIT00000.00268.fits')
```

## Chi-square values
Absolute values of $\chi^2$ is driven by DOF (~ N_pixels)

```python
colors = itertools.cycle(['k','b','r'])
for t, n in zip([dvc,ser,dvcexp], ['DVC', 'SER', 'DVCEXP']):
    scatter(t['DOF_'+n].data,
            t['CHISQ_'+n].data*t['DOF_'+n].data,
            label=n, facecolors=colors.next())

xlabel("DOF")
ylabel("$\chi^2$")
xscale('log')
yscale('log')
legend(loc='upper left')
```

## F-test

Model 1 is **nested** in model 2.

$$ F = \frac{ (\chi_1^2- \chi_2^2)/(dof_1 - dof_2) }{ \chi_2^2/dof_2 } \sim F(dof_1-dof_2, dof_2)$$

Test for the false-rejection probability $\alpha = 0.05$.

## Akaike Information Criterion (AIC, [Akaike 1974](http://adsabs.harvard.edu/abs/1974ITAC...19..716A))

$$ AIC = -2 \ln \mathcal{L}_{max} + 2k = \chi_{min}^2 + 2k $$
where $k$ is the number of free parameters

### "Jeffreys' scale"
- $\triangle AIC > 5$ : strong
- $\triangle AIC > 10$ : decisive

There are four extra free parameters ($\triangle k = 4$) in `DVCEXP` model ($I_e$, $R_e$, PA, b/a).

## Bayesian Information Criterion (BIC)

$$ BIC = -2 \ln \mathcal{L}_{max} + k \ln N $$
where $k$ is the number of free paramaters and $N$ is the number of data points

**Assumption** Data points are independent and equallity distributed.

[Liddle 2007](http://adsabs.harvard.edu/abs/2007MNRAS.377L..74L) as a starting reference for model selection

```python
redchi_dvc, dof_dvc = dvc['CHISQ_DVC'].data, dvc['DOF_DVC'].data
redchi_dvcexp, dof_dvcexp = \
    dvcexp['CHISQ_DVCEXP'].data, dvcexp['DOF_DVCEXP'].data
# calculate the F statistic
dvcexp_f = (redchi_dvc*dof_dvc - redchi_dvcexp*dof_dvcexp)\
            /(dof_dvc-dof_dvcexp)\
            /redchi_dvc
```

```python
alpha = 0.05
p_values = 1. - f.cdf(dvcexp_f,
                 dof_dvc - dof_dvcexp,
                 dof_dvcexp)
# reject the null hypothesis based on p-values
reject = p_values < alpha
print "N_reject / total =", count_nonzero(reject), "/", len(reject)

figure()
chi_diff = dvcexp['CHISQ_DVCEXP']*dvcexp['DOF_DVCEXP'] - \
            dvc['CHISQ_DVC']*dvc['DOF_DVC']
ylabel('delta $\chi^2$')
xlabel('DOF')
plot(dvc['DOF_DVC'][reject], chi_diff[reject], 'ko',
     label='reject')
plot(dvc['DOF_DVC'][~reject], chi_diff[~reject], 'r+', ms=10,
     label='accept')
yscale('symlog')
xscale('log')
legend(loc='lower left', title='F-test')

# AIC criterion lines
axhline(-13, c='b', ls='--')  # "strong"
axhline(-18, c='b', ls='-')  # "decisive"
text(1.08e4, -0.5e2, 'AIC decisive', color='b', fontsize=15)
```
